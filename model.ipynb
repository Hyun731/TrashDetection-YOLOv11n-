```python
from ultralytics import YOLO

def train_on_mac():
    model = YOLO("yolo11n.pt") 
    model.train(
        data="data.yaml",      
        epochs=50,
        imgsz=320,             
        device="mps",
        batch=16,              
        workers=4,            
        project="trash_project",
        name="on_device_model"
    )

    model.export(format="tflite", int8=True, data="data.yaml", imgsz=320)

if __name__ == "__main__":
    train_on_mac()
```
